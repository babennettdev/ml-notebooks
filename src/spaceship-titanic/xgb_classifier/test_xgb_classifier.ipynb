### Config

Set up directory variables.

```python
transformed_dataset_directory = "../transformed-data"
transformed_test_X_dataset_directory = f"{transformed_dataset_directory}/test_X.csv"

models_directory = "../models"
model_save_path = f"{models_directory}/xgb_classifier.json"

results_directory = "../results"
xgb_classifier_result_path = f"{results_directory}/xgb_boost_classifier_submission.csv"
```

### Dependencies

```python
%conda install pandas numpy matplotlib seaborn xgboost scikit-learn
```

```python
import pandas as pd

from xgboost import XGBClassifier
```

### Data Extraction

```python
X = pd.read_csv(transformed_test_X_dataset_directory)
X.head()
```

```python
passenger_ids = X['PassengerId'].values.reshape(-1, 1)
X.drop(['PassengerId'], axis=1, inplace=True)
```

```python
X.head()
```

```python
xgb_classifier = XGBClassifier()
xgb_classifier.load_model(model_save_path)
```

```python
predictions = xgb_classifier.predict(X)

passenger_ids = pd.DataFrame(passenger_ids, columns=["PassengerId"])
predictions = pd.DataFrame(predictions.astype('bool'), columns=["Transported"])

submission = pd.concat([passenger_ids, predictions ], axis=1)
submission.head(n=10)
```

```python
submission.to_csv(xgb_classifier_result_path, index=False)
```

#### Conclusions

Best result had an accuracy of 0.79471.
